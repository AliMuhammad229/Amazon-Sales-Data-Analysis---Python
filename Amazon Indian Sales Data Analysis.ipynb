# Group By

This function is use to group data based on one or more columns in a Dataframe

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)

sns.barplot(x='Size', y='Qty', data=S_Qty)
```

Note: From the above Graph yo can see that most of the quantities buys M-size in the sales

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=20, edgecolor = 'Black')
plt.xticks(rotation=90)
plt.show()
```

```python
#Checking B2B Data by using pie chart
B2B_Check = df['B2B'].value_counts()

#Plot the Pie Chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
plt.show()
```

```python
#Checking B2B Data by using pie chart
B2B_Check = df['B2B'].value_counts()

#Plot the Pie Chart
plt.title('Pie Chart')
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

Note: From the above Graph we can see that Maximum i.e. 99.2% of buyers are retailers and 0.8% are B2B buyers

```python
#Plot counts of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('Ship States')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plot counts of cities by state
#top 10 States
top_10_States = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_States.index)], x='ship-state')
plt.xlabel('Ship States')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=45)
plt.show()
```

Note: From the above Graph you can see that most of the buyers are Maharashtra state
